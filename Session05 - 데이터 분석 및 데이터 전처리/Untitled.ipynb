```python
import pandas as pd
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutDataStr = selloutData.astype ({"YEARWEEK":str})
```

```python
selloutData["YEAR"] = selloutDataStr.YEARWEEK.str[0:4]
selloutData["WEEK"] = selloutDataStr.YEARWEEK.str[4:6]
```

```python
selloutData
```

```python
newCustomerData = selloutData.loc[(selloutData.YEAR == "2015") & (selloutData.WEEK < "14")]
```

```python
groupKey = ["REGIONID"]
newCustomerData.groupby(by=groupKey)["QTY"].agg(['mean']).reset_index().sort_values('mean', ascending=False ,ignore_index=True)
```

##### kopo_channel_seasonality_new.csv 파일을 불러온 후
##### selloutData 변수에 담으세요 
##### 이후 QTY컬럼→실수(float), 이외컬럼→문자(str)로 변경하세요

##### kopo_channel_seasonality_new.csv 자료를 담은
##### selloutData 변수에서
##### QTY컬럼 음수(반품)인 경우 0, 양수인 경우 기존 QTY 값
##### 유지하는 로직을 적용하여 QTY_NEW 컬럼을 추가하세요

```python
import pandas as pd
```

```python
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutData["QTY_NEW"] = np.where(selloutData.QTY < 0, 0, selloutData.QTY)
```

```python
selloutData
```

##### selloutData 자료에서
##### YEAR, WEEK 컬럼을 생성하고 WEEK 가 52 이하인
##### 데이터만 조회한 후 refinedSelloutData 변수에 담으세요

```python
selloutDataStr = selloutData.astype ({"YEARWEEK":str})
```

```python
selloutDataStr.dtypes
```

```python
selloutData["YEAR"] = selloutDataStr.YEARWEEK.str[0:4]
```

```python
selloutData["WEEK"] = selloutDataStr.YEARWEEK.str[4:6]
```

```python
selloutData
```

```python
refinedSelloutData = selloutData[selloutData.WEEK.astype(int) <= 52]
```

```python
refinedSelloutData
```

### refinedSelloutData 에서 →
### 지역, 상품, 연주차 컬럼순으로 오름차순 정렬하여
### sortedData 변수에 담으세요

```python
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
```

```python
sortedData = refinedSelloutData.sort_values(sortKey)
```

```python
sortedData
```

### [지역, 상품, 연도 별 집계]
### sortedData 에서 지역, 상품, 연도 단위
### 판매량(QTY_NEW) 의 평균 연산 후
### groupData 변수에 담으세요
### 이후 컬럼명을 QTY_MEAN로 변경하세요

```python
groupKey = ["REGIONID","PRODUCT","YEAR"]
```

```python
groupData = sortedData.groupby(groupKey)[["QTY_NEW"]].agg("mean")
groupData = groupData.reset_index()

#특정 컬럼명 변경
groupData.rename(columns={'QTY_NEW': 'QTY_MEAN'})
```

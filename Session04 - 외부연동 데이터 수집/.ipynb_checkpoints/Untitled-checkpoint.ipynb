```python
import pandas as pd
from bs4 import BeautifulSoup 
import requests 
```

```python
targetUrl = "http://www.ssg.com/search.ssg?target=all&query=%EC%95%84%EC%9D%B4%ED%8F%B0"

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:71.0) Gecko/20100101 Firefox/71.0'}

resp=requests.get(targetUrl,headers=headers)

resp.encoding = "utf-8"

html = resp.text

htmlobj = BeautifulSoup(html, "html.parser")
```

```python
divtags = htmlobj.find(name = "div", attrs={"id":"area_searchItemList"})
```

```python
divtags
```

```python
emtags = divtags.findAll(name = "em")
```

```python
emtags
```

```python
eachEmtags = emtags[0]
```

```python
eachEmtagsText = eachEmtags.text
```

```python
eachEmtagsText
```

```python
divtags = htmlobj.findAll(name = "span", attrs={"style":"width:96.0%"})
```

```python
columnList = []
columnList1 =[]
columnList2 = []
```

```python
for i in range (2,len(emtags),7):
    eachEmtags = emtags[i]
    eachEmtagsText = eachEmtags.text
    columnList.append(eachEmtagsText)
for j in range (5,len(emtags),7):
    eachEmtags1 = emtags[j]
    eachEmtagsText1 = eachEmtags1.text
    columnList1.append(eachEmtagsText1)
for k in range (1,len(divtags)):
    eachEmtags2 = divtags[k]
    eachEmtagsText2 = eachEmtags2.text
    columnList2.append(eachEmtagsText2)
```

```python
finalFrame = pd.DataFrame(zip(columnList,columnList1,columnList2))
```

```python
finalFrame.columns = ["스마트폰은역시애플","그렇지만가격은사악","너희들의만족도"]
```
